```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from pdmpy import pdm
from astropy.timeseries import LombScargle
```

```python
period = 120
t = np.linspace(0, period, 1000)
y = norm.pdf(t, loc=7, scale=2)

n = 10
Y = np.tile(y, n)
T = np.linspace(0, period*10,len(Y))

Y = Y + np.random.normal(Y, scale=0.5)

plt.figure()
plt.plot(T, Y)

f_min = 1/(period*10)
f_max = 1/(period*0.1)

freq, theta = pdm(T, Y, delf=0.0001, f_min=f_min,f_max=f_max)

plt.figure()
plt.plot(1/freq, theta)
plt.axvline(period, color='red', alpha=0.1)
print(1/freq[np.nanargmin(theta)])

LS = LombScargle(T, Y)
f_, power = LS.autopower(minimum_frequency=f_min, maximum_frequency=f_max)
plt.figure()
plt.plot(1/f_, power)
plt.axvline(period, color='red', alpha=0.1)
print(1/f_[np.nanargmax(power)])
```
